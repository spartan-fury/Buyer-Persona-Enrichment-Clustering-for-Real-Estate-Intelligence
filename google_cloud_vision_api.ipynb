```python
# ┌───────────────────────────────────────────────────────────────────────┐
# │ 6️⃣ Extract WEB_DETECTION → pagesWithMatchingImages URLs              │
# └───────────────────────────────────────────────────────────────────────┘

PAGES=$(echo "${RESPONSE_JSON}" \
  | jq -r '.responses[0].webDetection.pagesWithMatchingImages[].url' \
  2>/dev/null)

if [[ -z "${PAGES}" ]]; then
  echo "⚠️ No pagesWithMatchingImages returned by webDetection."
else
  echo ""
  echo "=== All pagesWithMatchingImages URLs ==="
  echo "${PAGES}"
fi

```

```python
# ┌───────────────────────────────────────────────────────────────────────┐
# │ 7️⃣ Filter LinkedIn profile URLs                                      │
# └───────────────────────────────────────────────────────────────────────┘

echo ""
echo "=== LinkedIn profiles (linkedin.com/in/) ==="
echo "${PAGES}" \
  | grep -i "linkedin\.com/in/" \
  || echo "(none found matching linkedin.com/in/)"

```
